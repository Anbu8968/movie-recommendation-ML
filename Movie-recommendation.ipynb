```python
def make_generalist(x):
    gen=[]
    st=" "
    for i in x:
        if i.get("name")=="Science Fiction":
            scifi="Sci-Fi"
            gen.append(scifi)
        else:
            gen.append(i.get("name"))
    if len(gen)==0:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data["generalist"]=data["genres"].map(lambda x:make_generalist(x))
```

```python
data["generalist"]
```

```python
def get_actor1(x):
    casts=[]
    for i in x:
        casts.append(i.get("name"))
    if casts==[]:
        return np.NaN
    else:
        return casts[0]
```

```python
data["actor_1_name"]=data["cast"].map(lambda x:get_actor1(x))
```

```python
def get_actor2(x):
    casts=[]
    for i in x:
        casts.append(i.get("name"))
    if casts==[] or len(casts)<=1:
        return np.NaN
    else:
        return casts[1]
```

```python
data["actor_2_name"]=data["cast"].map(lambda x:get_actor2(x))
```

```python
def get_actor3(x):
    casts=[]
    for i in x:
        casts.append(i.get("name"))
    if casts==[] or len(casts)<=2:
        return np.NaN
    else:
        return casts[2]
```

```python
data["actor_3_names"]=data["cast"].map(lambda x:get_actor3(x))
```

```python
def get_directors(x):
    dt=[]
    st=" "
    for i in x:
        if i.get('job')=="Director":
            dt.append(i.get("name"))
    if dt==[]:
        return np.NaN
    else:
        return st.join(dt)
```

```python
data["director_name"]=data["crew"].map(lambda x:get_directors(x))
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if result:
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            np.NaN
    else:
        print('No search results found for query: {}'.format(x))
```
